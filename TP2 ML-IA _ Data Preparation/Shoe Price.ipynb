```python
# Type
shoes_dataset['Type'] = shoes_dataset['Type'].str.lower()  # Convertir toutes les valeurs de la colonne 'Type' en minuscules dans le dataset
```

```python
shoes_dataset['Type'].value_counts()  # Compter le nombre d'occurrences de chaque valeur unique dans la colonne "Type" du dataset
```

```python
shoes_dataset['Type'].value_counts().plot(kind='bar', legend='false', color='green')  # Créer un graphique à barres représentant le nombre d'occurrences de chaque valeur unique dans la colonne "Type" du dataset
plt.title('Type Counts')  # Définir le titre du graphique
plt.xlabel('Type names')  # Définir le nom de l'axe des abscisses
plt.ylabel('Counts')  # Définir le nom de l'axe des ordonnées
plt.show()  # Afficher le graphique
```

```python
# Définir une fonction pour ajouter un type aux chaussures
def add_type(inpt):
    # Vérifier si le type d'entrée est "casual", "fashion", "lifestyle", "slides" ou "retro"
    if inpt == "casual" or inpt == "fashion" or inpt == "lifestyle" or inpt == "slides" or inpt == "retro":
        return inpt  # Si oui, renvoyer le type d'entrée
    else:
        return "sport"  # Sinon, renvoyer "sport"

# Appliquer la fonction add_type à la colonne "Type" du dataset
shoes_dataset["Type"] = shoes_dataset["Type"].apply(add_type)
```

```python
shoes_dataset['Type'].value_counts() # Compter le nombre d'occurrences de chaque valeur unique dans la colonne 'Type' du dataset
```

```python
shoes_dataset['Type'].value_counts().plot(kind='bar', legend='false', color='green')  # Créer un graphique à barres représentant le nombre d'occurrences de chaque valeur unique dans la colonne 'Type' du dataset
plt.title('New Type Counts')  # Définir le titre du graphique
plt.xlabel('Type names')  # Définir le nom de l'axe des abscisses
plt.ylabel('Counts')  # Définir le nom de l'axe des ordonnées
plt.show()  # Afficher le graphique
```

```python
## Color
shoes_dataset['Color'] = shoes_dataset['Color'].str.lower() # Convertir les valeurs de la colonne 'Color' en minuscules
```

```python
# Fonction pour catégoriser les couleurs des chaussures
def add_Color(inpt):
    if inpt=='black' or inpt=='white' or inpt=='grey' or inpt=='black/white' or inpt=='pink' :
        return inpt
    else :
        return 'other'
# Appliquer la fonction add_color à la colonne 'Color' du dataset
shoes_dataset['Color'] = shoes_dataset['Color'].apply(add_Color)
```

```python
shoes_dataset['Color'].value_counts() # Compter le nombre d'occurrences de chaque catégorie de couleur
```

```python
shoes_dataset['Color'].value_counts().plot(kind='bar', legend='false', color='green')  # Créer un graphique à barres représentant le nombre d'occurrences de chaque couleur dans la colonne "Color" du dataset
plt.title('New Color Counts')  # Définir le titre du graphique
plt.xlabel('Color names')  # Définir le nom de l'axe des abscisses
plt.ylabel('Counts')  # Définir le nom de l'axe des ordonnées
plt.show()  # Afficher le graphique
```

```python
## Material
shoes_dataset['Material'] = shoes_dataset['Material'].str.lower() # Convertit les valeurs de la colonne "Material" en minuscules
```

```python
shoes_dataset['Material'].value_counts() # Compte le nombre d'occurrences de chaque matériau 
```

```python
shoes_dataset['Material'].value_counts().plot(kind='bar', legend='false', color='green')  # Créer un graphique à barres représentant le nombre d'occurrences de chaque matériau dans la colonne 'Material' du dataset
plt.title('Material Counts')  # Définir le titre du graphique
plt.xlabel('Material names')  # Définir le nom de l'axe des abscisses
plt.ylabel('Counts')  # Définir le nom de l'axe des ordonnées
plt.show()  # Afficher le graphique
```

```python
# Définir une fonction pour ajouter un matériau aux chaussures
def add_material(inpt):
    if inpt == 'mech' or inpt == 'leather' or inpt == 'canvas' or inpt == 'primeknit' or inpt == 'synthetic' or inpt == 'suede' or inpt == 'knit' or inpt == 'nylon':
        return inpt  # Si oui, renvoyer le matériau d'entrée
    else:
        return 'other'  # Sinon, renvoyer 'other'

# Appliquer la fonction add_material à la colonne 'Material' du dataset
shoes_dataset['Material'] = shoes_dataset['Material'].apply(add_material)
```

```python
shoes_dataset['Material'].value_counts() # Compter le nombre d'occurrences de chaque matériau
```

```python
shoes_dataset['Material'].value_counts().plot(kind='bar', legend='false', color='green')  # Créer un graphique à barres représentant le nombre d'occurrences de chaque matériau dans la colonne 'Material' du dataset
plt.title('New Material Counts')  # Définir le titre du graphique
plt.xlabel('Material names')  # Définir le nom de l'axe des abscisses
plt.ylabel('Counts')  # Définir le nom de l'axe des ordonnées
plt.show()  # Afficher le graphique
```
